### Step 1. Import the necessary libraries

```python
from datetime import datetime
import pandas as pd
import numpy as np

```

### Step 2. Create your time range (start and end variables). The start date should be 01/01/2015 and the end should today (whatever your today is)

```python
start_date = datetime(2015, 1, 1)

# Define the end date as today's day
end_date = datetime.today()

# Print the start and end dates
print("Start Date:", start_date)
print("End Date:", end_date)
```

### Step 3. Select the Apple, Tesla, Twitter, IBM, LinkedIn stocks symbols and assign them to a variable called stocks

```python
# Define the list of stock symbols
stocks = ["AAPL", "TSLA", "TWTR", "IBM", "LNKD"]

print("Stock Symbols:", stocks)
```

### Step 4. Read the data from google, assign to df and print it

```python
url = 'https://github.com/chenzeyu/stock_pred/raw/master/data/stock_AAPL.csv'

# Read the CSV file into a DataFrame, skipping problematic lines
AAPL = pd.read_csv(url , sep=',')

AAPL
```

```python
url = 'https://github.com/chenzeyu/stock_pred/raw/master/data/stock_TSLA.csv'

TSLA = pd.read_csv(url)
TSLA
```

```python
url = 'https://github.com/chenzeyu/stock_pred/raw/master/data/stock_TWTR.csv'

TWTR = pd.read_csv(url)

TWTR
```

# merge all three data and create a  one data frame name stocks

```python


# Combine the three DataFrames vertically
stocks_df = pd.concat([AAPL, TSLA, TWTR], ignore_index=True)


stocks_df

```

### Step 5.  What is the type of structure of df ?

### Step 7. Good, now we know  the data avaiable. Create a dataFrame called vol, with the Volume values.

```python
vol = stocks_df[['Volume']]

print(vol)
```

### Step 8. Aggregate the data of Volume to weekly
#### Hint: Be careful to not sum data from the same week of 2015 and other years.

### Step 9. Find all the volume traded in the year of 2015

```python

# Filter data for the year 2015
volume_2015 = stocks_df[stocks_df['Date'].dt.year == 2015]['Volume'].sum()

print("Total Volume Traded in 2015:", volume_2015)
```
